```python
import pandas as pd
import numpy as np
import csv
from ucimlrepo import fetch_ucirepo 
pd.set_option('display.max_columns', None)
```

## Cargar el DataSet Iris

```python
iris_areas = pd.read_csv("IrisAreas.csv")
iris_areas
```

### Separar cada clase en una matriz 

```python
setosa = iris_areas[iris_areas["species"] == "Iris-setosa"][["petal area", "sepal area"]]
versicolor = iris_areas[iris_areas["species"] == "Iris-versicolor"][["petal area", "sepal area"]]
virginica = iris_areas[iris_areas["species"] == "Iris-virginica"][["petal area", "sepal area"]]
virginica.head()
```

## Crear la función de disimilitud con la norma Euclidiana

```python
def disimilitud(x):
    n = len(x)
    p = len(x[0])
    dis = []
    for i in range(n):
        sum = 0
        for j in range(p):
           sum += x[i][j]
        dis.append(sum**(0.5))
    return dis
```

## Obtener la disimilitud entre el área del sépalo y pétalo de cada flor por cada especie

```python
setosa_dis = disimilitud(setosa.values.tolist())
versicolor_dis = disimilitud(versicolor.values.tolist())
virginica_dis = disimilitud(virginica.values.tolist())
```

### Escribir los resultados en un archivo csv

```python
with open("disimilitud.csv", "w", newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Setosa", "Versicolor", "Virginica"])
    for i in range(50):
        writer.writerow([setosa_dis[i], versicolor_dis[i], virginica_dis[i]])
```

## Pregunta 2 Umbral de Disimilitud

```python
umbral_setosa = np.mean(setosa_dis)
umbral_versicolor = np.mean(versicolor_dis)
umbral_virginica = np.mean(virginica_dis)
print(f"El umbral de setosa, versicolor y virginica respectivamente es: {umbral_setosa:.3f}, {umbral_versicolor:.3f}, {umbral_virginica:.3f}")
```

## Pregunta 3 Distancia entre los elementos

```python
  # fetch dataset 
iris_dataset = fetch_ucirepo(id=53) 
  
# data (as pandas dataframes) 
iris = iris_dataset.data.features 
iris['species'] = iris_dataset.data.targets 

iris
```

### Obtemos un dataset por cada clase de iris

```python
iris_setosa = iris[iris["species"] == "Iris-setosa"]
iris_versicolor = iris[iris["species"] == "Iris-versicolor"]
iris_virginica = iris[iris["species"] == "Iris-virginica"]
```

```python
#Quitamos la columna de species
iris_setosa.drop('species',axis=1,inplace=True)
iris_versicolor.drop('species',axis=1,inplace=True)
iris_virginica.drop('species',axis=1,inplace=True)
```

```python
#Lo volvemos matriz
iris_setosa = iris_setosa.values
iris_versicolor = iris_versicolor.values
iris_virginica = iris_virginica.values
```

```python
def distancia (x:np.ndarray,y: np.ndarray):
    rest= x-y
    ans=np.linalg.norm(rest)
    return ans
```

```python
dist_seto_versi= distancia(iris_setosa,iris_versicolor)
dist_seto_virgi= distancia(iris_setosa,iris_virginica)
dist_versi_virgi=distancia(iris_versicolor,iris_virginica)
dist_versi_seto=distancia(iris_versicolor,iris_setosa)
dist_virgi_seto=distancia(iris_virginica,iris_setosa)
dist_virgi_versi=distancia(iris_virginica,iris_versicolor)

print(f"La distancia entre setosa y versicolor es: {dist_seto_versi:.3f}")
print(f"La distancia entre setosa y virginica es: {dist_seto_virgi:.3f}")
print(f"La distancia entre versicolor y virginica es: {dist_versi_virgi:.3f}")
print(f"La distancia entre versicolor y setosa es: {dist_versi_seto:.3f}")
print(f"La distancia entre virginica y setosa es: {dist_virgi_seto:.3f}")
print(f"La distancia entre virginica y versicolor es: {dist_virgi_versi:.3f}")
```

# Pregunta 4 Producto interno

```python
#Definimos la función que nos da el producto interno de dos matrices

def prod_inter(x:np.ndarray,y:np.ndarray):
    C=x*y
    sum=0
    for i in range(len(C)):
        for j in range(len(C[0])):
            sum += C[i][j]
    return sum
```

```python
prod_inter_seto_versi= prod_inter(iris_setosa,iris_versicolor)
prod_inter_seto_virgi= prod_inter(iris_setosa,iris_virginica)
prod_inter_versi_virgi= prod_inter(iris_versicolor,iris_virginica)
prod_inter_versi_seto= prod_inter(iris_versicolor,iris_setosa)
prod_inter_virgi_seto= prod_inter(iris_virginica,iris_setosa)
prod_inter_virgi_versi= prod_inter(iris_virginica,iris_versicolor)

print(f"El producto interno entre setosa y versicolor es: {prod_inter_seto_versi:.3f}")
print(f"El producto interno entre setosa y virginica es: {prod_inter_seto_virgi:.3f}")
print(f"El producto interno entre versicolor y virginica es: {prod_inter_versi_virgi:.3f}")
print(f"El producto interno entre versicolor y setosa es: {prod_inter_versi_seto:.3f}")
print(f"El producto interno entre virginica y setosa es: {prod_inter_virgi_seto:.3f}")
print(f"El producto interno entre virginica y versicolor es: {prod_inter_virgi_versi:.3f}")
```
